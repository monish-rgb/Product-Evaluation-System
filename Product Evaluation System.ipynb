# ***Adding NAN values randomly ***

```python
nan_percent = {'ReviewTitle':0.001, 'ReviewBody':0.0002, 'ReviewStar':0.03,'Product':0.0001,'Price':0.012}

for col in df1:
    for i, row_value in df1[col].iteritems():
        if random.random() <= nan_percent[col]:
            df1[col][i] = np.nan
           

```

# **Analyzing Sentiment Of Reviews**

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
with_stpw['scores'] = with_stpw['ReviewBody'].apply(lambda ReviewBody: sid.polarity_scores(ReviewBody))
with_stpw.head()
```

```python
with_stpw['Total_Score']  = with_stpw['scores'].apply(lambda score_dict: score_dict['compound'])
with_stpw.head()
```

# **Arranging the dataset in descending order based on (Compound score) to find the most important sentence from the given data.**

```python
imp_sent=with_stpw.sort_values(by='Total_Score', ascending=False)
## arranging the compound column in descending order to find the best sentence.
imp_sent
```

# **Finding top positive sentence in the data.**

```python
with_stpw['positive']  = with_stpw['scores'].apply(lambda score_dict: score_dict['pos'])
```

```python
pos_sent=with_stpw.sort_values(by='positive', ascending=False)
## Arranging the dataframe by positive column in descending order to find the best postive sentence.
pos_sent
```

# **Finding Top Negative sentences in data.**

```python
with_stpw['negative']  = with_stpw['scores'].apply(lambda score_dict: score_dict['neg'])
```

```python
neg_sent=with_stpw.sort_values(by='negative', ascending=False)
## Arranging the dataframe by negative column in descending order to find the best negative sentence.
neg_sent
```

```python
with_stpw['Overall_Result']=with_stpw['Total_Score'].apply(lambda c:'Positive' if(c>=0) else 'Negative')
with_stpw.head()
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(15,8)) 
X = models['Models']
Traning = models['Training_Accuracy']
Testing=models['Testing_Accuracy']
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Traning, 0.4, label = 'Traning_Accuracy')
plt.bar(X_axis + 0.2, Testing, 0.4, label = 'Testing_Accuracy')

plt.xticks(X_axis, X) 
plt.xlabel("Models",fontweight='bold')
plt.ylabel("Accuracy",fontweight='bold')
plt.title("$Accuracy$ $Comparision$ $Of$ $Models$")
plt.legend(loc='upper right',bbox_to_anchor=(1.2, 1.0))
plt.show()
```

```python
plt.figure(figsize=(15,7)) 
X = models['Models']
Precision = models['Precision']
Recall =models['Recall']
F1score=models['F1_Score']
  
X_axis = np.arange(len(X))
barWidth = 0.25
# Set position of bar on X axis
pos1 = np.arange(len(X))
pos2 = [x + barWidth for x in pos1]
pos3 = [x + barWidth for x in pos2]
  
plt.bar(pos1,Precision,width=barWidth, label = 'Precision')
plt.bar(pos2,Recall,width=barWidth, label = 'Recall')
plt.bar(pos3,F1score,width=barWidth, label = 'F1_Score')
  
plt.xticks(X_axis, X)
plt.xlabel("Models",fontweight='bold')
plt.ylabel("Accuracy",fontweight='bold')

plt.title("$Accuracy$ $Comparision$ $Of$ $Models$")
plt.legend(loc='upper right',bbox_to_anchor=(1.2, 1.0))
plt.show()
```

```python
plt.figure(figsize=(15,6))
x=models['Models']
y=models['Wrongly_Predicted_values']
ax = plt.axes()
ax.set_facecolor("white")
plt.bar(x,y,color = '#94E413', width=0.5)
plt.xlabel('Models',fontweight='bold')
plt.ylabel('No_Of_Wrong_Predictions',fontweight='bold')
plt.title ('$Wrongly$ $Predicted$ $values$ $For$ $Each$ $Model$')
plt.show()
```
